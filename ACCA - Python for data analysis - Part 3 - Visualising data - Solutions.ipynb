Grouping is a fundamental part of many data analyses and metric calculations:
- "What's the total sales **BY DEPARTMENT**?"
- "What's the Average Revenue **Per User**?"
- "**Who** claimed the most valuable expense?"
- "**Which year** had the lowest total VAT payments?"

Grouping must always be paired with an **aggregation** function. The above questions would be answered as follows:
- `df.groupby("department").sales.sum()`
- `df.groupby("user").revenue.mean()`
- `df.groupby("employee").expense_value.max()`
- `df.groupby("year").vat_paid.sum().sort_values()`

````python
# What's the average age by destination?
orders.groupby('Destination Country').Age.mean()
````

````python
# You may want to add .reset_index() on the end of this to turn it back into a DataFrame.
orders.groupby('Destination Country').Age.mean().reset_index()
````

````python
# Let's sort as well...
(
    orders
    .groupby('Destination Country')
    .Age
    .mean()
    .reset_index()
    .sort_values('Age')
)
````

````python
# Let's break this down further by destination country AND city.

# We will use brackets to make our code easier to read according to the
# "Modern Pandas" style guide by Tom Augsperger.

(
    orders
    .groupby(['Destination Country', 'Destination City'])
    [['Age', 'Hotel Rating']]
    .mean()
    .reset_index()
    .sort_values('Hotel Rating', ascending=False)
    .head()
)
````

````python
# To count rows in pandas, use .size()

# Which country MADE the most bookings?
orders.groupby('Origin Country').size().reset_index()
````

> ### ðŸš© Exercises
> - Which destination city do people stay the longest on average? (_Tip: You will need to use square brackets to reference the `No. Of Days` column_)
> - What is the most popular destination city in China? (_Tip: You will need to filter to Chinese destinations first and use the `.size()` method to count rows by city._)

````python
# âœ… SOLUTION

# Which destination city do people stay the longest on average?
(
    orders
    .groupby(['Destination Country', 'Destination City'])
    ['No. Of Days']
    .mean()
    .sort_values(ascending=False)
    .reset_index()
)
````

````python
# âœ… SOLUTION

# What is the most popular destination city in China?
(
    orders
    .query("`Destination Country` == 'China'")
    .groupby("Destination City")
    .size()
    .sort_values(ascending=False)
    .reset_index()
    .rename(columns={0: 'Total Orders', 'Destination City': 'City'})
)
````

Seaborn is a library for making statistical graphics that can be somewhat fiddly using either matplotlib or pandas. It uses matplotlib under-the-hood and is designed to work well with dataframes.

To learn more, read _[An Introduction to Seaborn](https://seaborn.pydata.org/introduction.html)_, take a look at the [example gallery](https://seaborn.pydata.org/examples/index.html) or check out the [official tutorial](https://seaborn.pydata.org/tutorial.html).

````python
import seaborn as sns
````

````python
# Let's visualise this again using Seaborn. Note how we pass in the dataframe as
# an argument and then simply reference the dataframe's columns as strings.

sns.barplot(
    x='No. Of People',   # plot this on the x-axis
    y='Origin Country',  # plot this on the y-axis
    estimator=sum,       # calculate totals
    orient='h',          # make it a horizontal barplot
    data=orders          # Seaborn needs to know which dataframe to look at
);
````

````python
# By simply adding "hue='Gender'" we can drill down and explore how the
# booking count varies by both country of origin AND gender

sns.barplot(
    x='No. Of People',
    y='Origin Country',
    hue='Gender',        # colour code by gender
    estimator=sum,
    orient='h',
    data=orders
);

# Adding the semicolon ; at the end "suppresses" the Jupyter output, i.e.
# we no longer see "<AxesSubplot:xlabel='No. Of People', ylabel='Origin Country'>"
````

````python
# Let's add a title to this plot. We do this by assigning the plot to a variable
# and then using the .set() method on the variable.

my_plot = sns.barplot(x='No. Of People', y='Origin Country', hue='Gender',
                      estimator=sum, orient='h', data=orders)

my_plot.set(xlabel='Number of People', ylabel='Country', title="Total individuals by country of origin");
````

### How do we add `Booking Month`?
Let's add booking month. We could also extract:
- month with `DATAFRAME.COLUMN.dt.month_name()`
- day with `DATAFRAME.COLUMN.dt.day_name()`
- day of year with `DATAFRAME.COLUMN.dt.day_of_year`

and many more!

If it doesn't work for you, you may need to convert your column of interest to "datetime" format, using e.g.

```python
DATAFRAME['COLUMN'] = pd.to_datetime(DATAFRAME.COLUMN)
```

Find out more on the [docs page for the `pd.to_datetime()` function](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.to_datetime.html).

````python
orders['Booking Month'] = orders['Date of Booking'].dt.month_name()
````

````python
# Booking Month is now on the right-hand side of the dataframe.
orders.head(3)
````
